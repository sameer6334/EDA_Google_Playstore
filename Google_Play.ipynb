```python
def size_process(item):
    if str(item)[-1] == "M":
        result = float(str(item).replace("M",""))
        result = result*1024
        return result

    elif str(item)[-1] == "K":
        result = float(str(item).replace("K",""))
        return result
    else:
        return str(np.nan)
```

```python
def clean_version(version):
    if not isinstance(version, str) or version.strip() == "":  # Handle NaN and empty values
        return np.nan  

    version = version.strip()  # Remove extra spaces

    # Handle version ranges: "4.0.3 - 7.1.1"
    if "-" in version:
        versions = version.split("-")  # Split into two parts
        versions = [clean_version(v) for v in versions]  # Recursively clean each part
        return max(versions)  # Keep the highest version

    # Remove third decimal place if present (e.g., "4.0.3" -> "4.0")
    parts = version.split(".")
    if len(parts) > 2:
        version = parts[0] + "." + parts[1]  # Keep only the first two parts

    try:
        return float(version)  # Convert to float
    except ValueError:
        return np.nan  # If conversion fails, return NaN
```

```python
df_copy["Android Ver"] = df_copy["Android Ver"].apply(clean_version)
```
